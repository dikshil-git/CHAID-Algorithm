```python
# Step 1: Importing Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
# Note: For strict CHAID implementation, 'CHAID' package can be installed via pip install CHAID
# Alternatively, we can use DecisionTreeClassifier configured with categorical discretization or an explicit CHAID wrapper.
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.preprocessing import OrdinalEncoder

# Step 2: Loading the Dataset
# Creating an appropriate categorical dataset for CHAID demonstration (e.g., purchasing decision based on demographics)
data = pd.DataFrame({
    'Age': ['Youth', 'Youth', 'Senior', 'Middle', 'Senior', 'Middle', 'Youth', 'Senior'],
    'Income': ['High', 'Low', 'Medium', 'Medium', 'High', 'Low', 'Low', 'High'],
    'Student': ['No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'Yes'],
    'Credit_Rating': ['Fair', 'Excellent', 'Fair', 'Fair', 'Excellent', 'Fair', 'Excellent', 'Fair'],
    'Buys_Computer': ['No', 'Yes', 'Yes', 'Yes', 'No', 'No', 'Yes', 'Yes']
})

# Step 3: Understanding the Dataset
print("--- Step 3: Understanding the Dataset ---")
print(data.head())
print("\nDataset Info:")
print(data.info())

# Step 4: Passing the Data Frame (Encoding categorical variables for modeling)
encoder = OrdinalEncoder()
feature_cols = ['Age', 'Income', 'Student', 'Credit_Rating']
X = encoder.fit_transform(data[feature_cols])
y = data['Buys_Computer']

# Training the decision tree model (acting as proxy/demonstration for interaction detection)
clf = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=42)
clf.fit(X, y)

# Step 5: Prediction using instance – Passing Index (Index 0 from original data)
print("\n--- Step 5: Prediction using instance (Index 0) ---")
sample_index = 0
instance_index_data = X[sample_index].reshape(1, -1)
pred_index = clf.predict(instance_index_data)
print(f"Prediction for row index {sample_index}: {pred_index[0]}")

# Step 6: Prediction using instance (A custom new instance vector)
print("\n--- Step 6: Prediction using custom instance ---")
# Example: [Youth, High, No, Fair] -> encoded appropriately
new_instance = encoder.transform([['Youth', 'High', 'No', 'Fair']])
pred_instance = clf.predict(new_instance)
print(f"Prediction for custom instance: {pred_instance[0]}")

# Step 7: Prediction using data values (Batch prediction)
print("\n--- Step 7: Prediction using data values ---")
batch_samples = encoder.transform([
    ['Senior', 'Low', 'Yes', 'Excellent'],
    ['Middle', 'Medium', 'No', 'Fair']
])
batch_preds = clf.predict(batch_samples)
print("Batch Predictions:", list(batch_preds))

# Step 8: Decision Rule
print("\n--- Step 8: Decision Rules ---")
tree_rules = export_text(clf, feature_names=feature_cols)
print(tree_rules)
```
